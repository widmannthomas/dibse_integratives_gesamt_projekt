```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,10)
```

# read the measurements file

```python
data_frame = pd.read_csv("udp_tcp.csv", sep=" ", names=["time_ms", "current_mA"])
data_frame.describe()
```

# WiFi connect cycle

```python
plt.plot(data_frame[10000:20000].current_mA, label="Connection cycle [mA]")
plt.legend()
plt.show()
```

# Deep sleep power consumption

```python
plt.plot(data_frame.current_mA[100:400], label="deep sleep [mA]")
plt.plot(data_frame.current_mA[100:400].rolling(30).mean(), label="moving average [mA]")
plt.legend()
plt.show()
```

# Wake up phase

```python
plt.plot(data_frame.current_mA[2900:3100], label="deep sleep [mA]")
plt.legend()
plt.show()
```
